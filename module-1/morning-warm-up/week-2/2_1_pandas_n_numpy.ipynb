##### Use a for loop to find all rows where column 3 entries have values greater than 2 and replace them with NaN values 
##### Hint: Use NumPy

```python
import numpy as np

for i in df.index:
    if df.iloc[i,3] > 2:
        df.iloc[i,3] = np.nan

df.head(5)
df.info()
       
```
